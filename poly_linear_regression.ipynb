```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Employee.csv")
```

```python
df.head()
```

```python
x = np.array(df.iloc[:,1:2])
y = np.array(df.iloc[:,2:])
y
```

```python
plt.scatter(x,y)
plt.show()
```

### Lets try to fit Simple linear relation model


```python
from sklearn.linear_model import LinearRegression

model1 = LinearRegression()
model1.fit(x,y)
predicted_mod1 = model1.predict(x)
predicted_mod1
```

```python
plt.scatter(x,y,color = "green")
plt.plot(x,predicted_mod1,color = "red")
plt.xlabel("Input")
plt.ylabel("Prediction")
plt.show()
```

### Polynomial Linear regression

```python
# Before fitting we have to make sure the values are polynomial
# We can do it manually , however we will use a sklearn library

from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree=2)
x_poly = poly_reg.fit_transform(x)
```

```python
poly_linear_reg = LinearRegression()
poly_linear_reg.fit(x_poly,y)
poly_prediction = poly_linear_reg.predict(x_poly)
```

```python
plt.scatter(x,y ,color = "red")
plt.plot(x,poly_prediction, color = "blue")
```

### Lets fit it for degree 3

```python
poly_reg_3 = PolynomialFeatures(degree= 3)
poly_vars = poly_reg_3.fit_transform(x)
```

```python
poly_linear_reg.fit(poly_vars,y)
poly_prediction = poly_linear_reg.predict(poly_vars)
```

```python
plt.scatter(x,y ,color = "red")
plt.plot(x,poly_prediction, color = "blue")
plt.show()
```
